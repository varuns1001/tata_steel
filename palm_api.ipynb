```python

from google.cloud import aiplatform
aiplatform.init(project = 'tsl-ai-prod', location = "us-central1")

# Automatically restart kernel after installs so that your environment can access the new packages

from vertexai.preview.language_models import TextGenerationModel,\
                                            ChatModel,\
                                            InputOutputTextPair,\
                                            TextEmbeddingModel

from IPython.display import display, Markdown
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
generation_model = TextGenerationModel.from_pretrained("text-bison@001")

# import vertexai
# vertexai.init(project = 'tsl-datalake', location = "asia-south1")
```

```python
def summarizer(text):
  prompt = f"""
          Summarize this following news

          {text}
  """
  response = generation_model.predict(
    prompt,
    temperature=0.7,
    max_output_tokens=1024,
    top_p=0.8
  )
  return response.text
```

```python
def locater(text):
  prompt = f"""
          Find all the locations from the following text and return a json.

          {text}
  """
  response = generation_model.predict(
    prompt,
    temperature=0.7,
    max_output_tokens=1024,
    top_p=0.8
  )
  return response.text
```

```python
tata_steel = pd.read_csv('tata steel news.csv')
tata_steel
```

```python
tata_steel_subset = tata_steel.loc[0:99]
tata_steel_subset
```

```python
tata_steel_subset.columns
```

```python
tata_steel_subset['summary'] = tata_steel_subset['Article'].apply(summarizer)
tata_steel_subset
```

```python
tata_steel_subset['location'] = tata_steel_subset['Article'].apply(locater)
```
